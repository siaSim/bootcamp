```python
# 전화번호 검색 프로그램 * 딕셔너리 이용

addr = {
    '박지민': '010-1111-1234', '전정국': '010-2222-1234',
    '민윤기': '010-3333-1234', '정호석': '010-4444-1234',
    '김태형': '010-5555-1234', '김석진': '010-6666-1234'
}

while True:
    answer = int(input("메뉴 선택(1:추가, 2:검색, 0:종료)? "))

    if answer == 1: #추가기능
        add = input("친구 이름 입력 >> ")
        addadr = input("전화번호 입력 >> ")
        addr[add] = addadr

    elif answer == 2: #검색기능
        print("검색가능한 친구들 :", list(addr.keys()))
        search = input("친구이름 입력 >> ")

        if search in addr: 
            print(f"{search}의 연락처 : {addr[search]}")
        else:
            print("등록되지 않은 친구입니다.")

    elif answer == 0: #종료
        print("프로그램을 종료합니다!")
        break

    else:
        print("메뉴 선택 오류입니다. 다시 선택하세요.")
```

```python
# 세 개의 숫자를 입력받아 가장 큰수를 출력하는 num_max 함수를 정의
# max 함수 쓴 version
'''
def num_max(a, b, c):
    return max(a, b, c)

a, b, c = map(int, input().split())
print(num_max(a, b, c))
'''
# max 함수 안 쓴 version
def num_max(a, b, c):
    biggest = a

    if b > biggest:
        biggest = b

    if c > biggest:
        biggest = c

    return biggest

a, b, c = map(int, input().split())
print(num_max(a, b, c))
```

```python
# 숫자들의 배열을 입력하면 가장 작은 값을 구하는 함수
def num_min(arr):
    small = arr[0]
    
    for i in arr:
        if i < small:
            small = i
    return small
    
numbers = input().split()

num_list = []
for i in numbers:
    num_list.append(int(i))

print(num_min(num_list))
```

```python
# 윤년을 확인하는 함수 만들기
# 4로 나누어 떨어지는 해는 윤년
# 100으로 나누어 떨어지는 해는 윤년이 아님
# 400으로 나누어 떨어지는 해는 윤년

def isLeapYear(year):
    if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
        return True
    else:
        return False
    
print(f'2022년은 윤년일까? {isLeapYear(2022)}')
print(f'2010년은 윤년일까? {isLeapYear(2010)}')
print(f'2000년은 윤년일까? {isLeapYear(2000)}')
print(f'1900년은 윤년일까? {isLeapYear(1900)}')

'''
또 다른 방법
def isLeapYear(year):
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False
'''
```

```python
def add_many(*args):
    print(args)
    # 입력 정수들의 합을 리턴
    sum = 0
    for i in args:
        sum += i
    return sum
```

```python
add_many(1,2,3)
```

```python
add_many(1,2,3,4,5)
```

```python
# 여러 개의 입력값을 받는 함수 만들기

def add_mul(*args):
    if args[0] == 'add':
        result = 0
        for i in args[1:]:
            result += i
    elif args[0] == 'mul':
        result = 1
        for i in args[1:]:
            result *= i

    return result

print(add_mul('add', 1, 2, 3))
print(add_mul('mul', 1, 2, 3, 4, 5))
```
