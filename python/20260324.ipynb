```python
# 미션
marks = [90, 25, 90, 45, 80]
# 60점 이상 합격, 60점 미만 불합격

num = 0
for i in marks:
    num += 1
    if i >= 60:
        print(f'{num}번 학생 {i}점으로 합격')
    else:
        print(f'{num}번 학생 {i}점으로 불합격')
```

```python
# 도전미션
# 리스트 numbers 각 원소는 숫자 0부터 9까지로 이루어져 있습니다. 이때, numbers에서 연속적으로 나타나는 숫자는 하나만 남기고 전부 제거하려고 합니다.
# 단, 제거된 후 남은 수들은 numbers의 원소의 순서를 유지해야 합니다.

# numbers = [1,1,3,3,0,1,1,] -> [1,3,0,1]
# numbers = [4,4,4,3,3]->[4,3]

'''
numbers = [1,1,3,3,0,1,1]
result = []

for i in numbers:
    if len(result) == 0:
        result.append(i)
    elif result[-1] != i:
        result.append(i)
print(result)
'''
numbers = [1,1,3,3,0,1,1]
answer = []
comp = numbers[0]
answer.append(comp)

for i in numbers:
    if comp != i:
        comp = i
        answer.append(comp)
answer
```

```python
# 전화번호 검색 프로그램 * 딕셔너리 이용

addr = {
    '박지민': '010-1111-1234', '전정국': '010-2222-1234',
    '민윤기': '010-3333-1234', '정호석': '010-4444-1234',
    '김태형': '010-5555-1234', '김석진': '010-6666-1234'
}

while True:
    answer = int(input("메뉴 선택(1:추가, 2:검색, 0:종료)? "))

    if answer == 1:
        add = input("친구 이름 입력 >> ")
        addadr = input("전화번호 입력 >> ")
        addr[add] = addadr

    elif answer == 2:
        print("검색가능한 친구들 :", list(addr.keys()))
        search = input("친구이름 입력 >> ")

        if search in addr:
            print(f"{search}의 연락처 : {addr[search]}")
        else:
            print("등록되지 않은 친구입니다.")

    elif answer == 0:
        print("프로그램을 종료합니다!")
        break

    else:
        print("메뉴 선택 오류입니다. 다시 선택하세요.")
```
