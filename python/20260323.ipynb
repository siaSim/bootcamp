```python
# 미션 : 사용자로부터 두 수를 입력 받아서, 합 차 곱 나눗셈을 출력해주세요.
num1 = int(input('첫 번째 숫자를 입력하세요: '))
num2 = int(input('두 번째 숫자를 입력하세요: '))

print('합:', num1 + num2)
print('차:', num1 - num2)
print('곱:', num1 * num2)
print('나눗셈:', num1 / num2)
```

```python
# //, % 연산자 사용 연습
n = 1234
d1 = n // 1000
d2 = (n // 100)%10
d3 = (n // 10)%10
d4=n%10
print(d1, d2, d3, d4)
```

```python
# 변수 a에 저장된 값의 b자리 값을 계산하여 변수 c에 저장하는 코드를 설계하라
# 예를 들어 a에 12345가 저장되어 있고, b가 1일 경우 첫번째 자리값인 4가 c에 저장된다.
# a가 1234일 때 b가 4일 경우는 4번째 자리값인 1이 c에 저장되어야한다.

a = 12345
b = 1
c = (a % (10 ** b)) // (10 ** (b - 1))
print(c)

#1234//10**(b-1)%10
```
