```python
# 둘다 90점 이상이면 둘다 성공
# 영어는 90점 이상인데 수학은 90점 미만이면 좀 더 노력
# 둘다 90점 미만이면 많이 노력


eng = int(input('영어 점수를 입력하세요: '))
math = int(input('수학 점수를 입력하세요: '))

if eng >= 90 and math >= 90:
    print('둘 다 성공!')
elif eng >= 90 and math < 90:
    print('좀 더 노력')
elif eng < 90 and math >= 90:
    print('좀 더 노력')
else:
    print('많이 노력')
```

```python
# 미션 : 성적처리 프로그램
# 90점 이상 100점 미만 A
# 80점 이상 90점 미만 B
# 70점 이상 80점 미만 C
# 60점 이상 70점 미만 D
# 60점 미만 F

score = int(input('점수를 입력하세요 :'))

if score >=90
    print('A')
elif score >= 80
    print('B')
elif score >= 70
    print('C')
elif score >= 60
    print('D')
else:
    print('F')
```

## 반복문

```python
for, while(while True)
for 변수 in 이터러블:
    반복할문장
```

```python
for i in range(1, 5+1, 2):
    print(f'{i}th Hello Python ')
```

```python
# 카운트다운 10, 9, 8, 7, 6 ...

for i in range(10, 0, -1):
    print(f"{i}th Hello Python")
```

```python
# for 반복문과 range() 함수를 사용해서 1부터 10까지의 합을 출력해주세요.
sum = 0

for i in range(1, 11):
    sum = sum +i

print(sum)
    
```

```python
# for 반목문과 range() 함수를 사용해서 1부터 10까지의 곱을 출력해주세요.
sum=1

for i in range(1, 11):
    sum = sum * i

print(sum)
    
```

```python
# for 반복문과 range() 함수를 사용해서 구구단을 출력
for i in range(2, 10):
    print(i, "단")
    for j in range(1, 10):
        print(i, "*", j, "=", i * j, end='\t')
    print()
```

```python
#2단 ~ 9단 출력

for j in range(1, 10):
    for i in range(2, 10):
        print(i, '*', j, '=', i*j, end='\t')
    print()
```

```python
while 조건:
    반복할문장
```

```python
i=0

while i<5:
    print(f"{i}th Hello Python")
    i += 1
```

```python
# 기본미션 : while 반복문을 1부터 사용자로부터 입력받은 수까지의 합을 출력

n = int(input("수입력: "))
sum = 0
i = 1

while i <= n:
    sum = sum + i
    i += 1

print(sum)
```

```python
while True:
    실행할명령
    
break, continue
```

```python
# 1부터 10까지의 합을 구하되, 합이 20 이상이 되는 때의 정수를 구하시오.
hap = 0

for i in range(1, 11):
    hap+=i
    if hap >= 20:
        print(f"정수: {i}, 합계: {hap}")
        break
hap
```

```python
# up and down 게임 만들기 1~100 사이 맞추는 게임
import random

answer = random.randint(1, 100)
count = 0

while True:
    number = int(input("숫자 : "))
    count += 1

    if number > answer:
        print("Down")
    elif number < answer:
        print("UP")
    else:
        print("Correct!")
        print(count, "번 만에 맞췄습니다.")
        break
```
