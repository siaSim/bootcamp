```python
class Person:
    population = 0 # 클래스 변수(공유)
    def __init__(self, name, age):
        #초기화루틴
        self.name = name
        self.age = age
        Person.population += 1
        
        
    def walk(self):
        print(f'{self.name}이 걸어가요')
    
```

```python
person01 = Person('박지민', 31)
```

```python
print(type(person01))
```

```python
Person.population
```

```python
person01.age
```

```python
person01.walk()
```

```python
person02 = Person('전정국', 30)
```

```python
person02.walk()
```

```python
#상속

class NewPerson(Person):
    #기능확장
    def fly(self):
        print(f'{self.name}이 날고있어요!!!')
    #메서드 오버라이딩
    def walk(self):
        print(f'{self.name}이 아주 빠르게 걸어가요')
    #수영기능
    def swim(self):
        print(f'{self.name}이 헤엄치고 있어요')
```

```python
person03=NewPerson('정호석', 33)
```

```python
person03.age
```

```python
person03.walk()
```

```python
person03.fly()
```

```python
person03.swim()
```
